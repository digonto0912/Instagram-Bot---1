```python
## function for search_open the profile

def search_open(driver, profile):
    
    driver.get("https://www.instagram.com/")
    
    open_ = True
    
    wait = WebDriverWait(driver, 10)       ## Explicit Wait
    
    try:
        search_path = "//div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div/div/div[2]/div[2]/span/div/a"
        search_btn = wait.until(EC.presence_of_element_located((By.XPATH,search_path)))
        search_btn.click()
        
        ## search 'So Delhi' 
        search_path = "//div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div[2]/div/div/div[2]/div/div/div[1]/div/div/input"
        search_bar = wait.until(EC.presence_of_element_located((By.XPATH,search_path)))
        search_bar.clear()
        search_bar.send_keys(profile)
        
        time.sleep(3)

        ## open
        path = '//div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div[2]/div/div/div[2]/div/div/div[2]/div/a[1]'
        btn = wait.until(EC.element_to_be_clickable((By.XPATH,path)))
        btn.click()
        
    except Exception:
        open_ = False
        
    return open_
```

## Profile Closing Function

```python
## function for closing profile or returning home page

def close_profile(driver):
    global on
    
    wait = WebDriverWait(driver, 10)
    home = wait.until(EC.element_to_be_clickable((By.XPATH, '//div/div/div[2]/div/div/div[1]/div[2]/div/div[1]/div/div/div/div/div[2]/div[1]/div/span/div/a')))
    # home.click()
    # we are Useing JavaScript click because normal click doesn't work in this case
    driver.execute_script("arguments[0].click();", home)
    
    on = 0
```

```python
## function for log in

def remove_notification(driver, wait):
    ## Notifination popup setting up not now, if occur
    time.sleep(3)
    
    driver.get(site)
    
    try:
        popup_path = '/html/body/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[3]/button[2]'
#         /html/body/div[4]/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[3]/button[2]
        not_now_btn = wait.until(EC.visibility_of_element_located((By.XPATH, popup_path)))
        not_now_btn.click()
    except TimeoutException:
        print("Timeout while waiting for the notification popup.")
    except NoSuchElementException:
        print("Notification popup not found.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

def log_in_to_account(driver, username, password):
    
    time.sleep(3)
    
    login = True
    
    wait = WebDriverWait(driver, 10)  ## explicit wait
    
    try:

        ## enter username
        username_ = wait.until(EC.presence_of_element_located((By.XPATH,"//*[@id='loginForm']/div/div[1]/div/label/input")))
        username_.send_keys(username)

        ## enter password
        password_ = wait.until(EC.presence_of_element_located((By.XPATH,"//*[@id='loginForm']/div/div[2]/div/label/input")))
        
        password_.send_keys(password)

        ## click login
        log_in_path = '/html/body/div[2]/div/div/div[2]/div/div/div[1]/section/main/article/div[2]/div[1]/div[2]/form/div/div[3]/button'
        log_in = wait.until(EC.presence_of_element_located((By.XPATH,log_in_path)))
        log_in.submit()

        time.sleep(3)  ## waiting for website to switch
        
    except Exception:
        login = False
        
    if login:
        remove_notification(driver, wait)
    
    return login

## 1

login = log_in_to_account(driver, username, password)
if login:
    print("log in successfully")
else:
    print("log in unseccussful, give one more try")
```

## 4. Follow / Unfollow

```python
## function to follow a given handle/profile

def follow(driver, profile):
    
    previously_follow = False
    
    wait = WebDriverWait(driver, 10)       ## Explicit Wait
    
    ## opening profile
    search_open(driver, profile)
    
    ## following handle
    path = '//div/div/div[2]/div/div/div[1]/div[2]/div/div[2]/section/main/div/header/section[2]/div/div/div[2]/div/div[1]/button/div/div'
    btn = wait.until(EC.presence_of_element_located((By.XPATH, path)))
    print(btn.get_attribute('innerHTML'))
    
    ## Checking if i am following or not
    if(btn.get_attribute('innerHTML') == "Follow"):
        btn = wait.until(EC.element_to_be_clickable((By.XPATH, path)))
        btn.click()
        
        time.sleep(3)
    else:
        previously_follow = True
        
    ## Closing profile
    close_profile(driver)
    
    return previously_follow    



## function to unfollow a given profile

def unfollow(driver, profile):
    
    previously_unfollow = False
    
    wait = WebDriverWait(driver, 10)       ## Explicit Wait
    
    ## opening profile
    search_open(driver, profile)
    
    ## unfollowing handle
    path = '//div/div/div[2]/div/div/div[1]/div[2]/div/div[2]/section/main/div/header/section[2]/div/div/div[2]/div/div[1]/button/div/div'
    btn = wait.until(EC.presence_of_element_located((By.XPATH,path)))
    
    ## Checking if i am unfollowing or not
    if(btn.get_attribute('innerHTML') != "Follow"):
        btn = wait.until(EC.element_to_be_clickable((By.XPATH,path)))
        btn.click()
        
        ## unfollowing
        time.sleep(2)
        path = '//div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div/div[8]/div[1]'
        btn = wait.until(EC.element_to_be_clickable((By.XPATH,path)))
        btn.click()
    else:
        previously_unfollow = True
        
    time.sleep(5)
    
    ## Closing profile
    close_profile(driver)
    
    return previously_unfollow  



## 4

profile = "So Delhi"        ## can be changed

## follow
previously_follow = follow(driver, profile)
if(previously_follow):
    print("Already Following")
else:
    print("Start Following")
    
    
time.sleep(4)    


## unfollow
previously_unfollow = unfollow(driver, profile)
if(previously_unfollow):
    print("Already not Following")
else:
    print("Start Unfollowing")
    
```

```python
# Here I try to solve the post-collecting problem, using by beautiful soup 
# but Instagram has some personal way of showing posts, that's why we never use this approach
# learn more about this problem -> " https://youtu.be/96PssSyB5Io?si=2PmoBtPC-jMA2sm0 "
```

```python
## function to liking the first no_of_post of a profile
def like(driver, profile, no_of_post):
    
    previously_like = False
    
    wait = WebDriverWait(driver, 10)       ## Explicit Wait
    
    ## opening profile
    search_open(driver, profile)
    
    ## Extracting The No_of_posts
    path = "//div[contains(@class,'x1lliihq x1n2onr6 xh8yej')]"
    btn = wait.until(EC.presence_of_all_elements_located((By.XPATH,path)))
    while len(btn) < no_of_post:
        driver.execute_script('window.scrollBy(0, 1000);')         ## Scrolling
        btn = wait.until(EC.presence_of_all_elements_located((By.XPATH,path)))
        
    print("number of posts extracted: ", len(btn))
    
    cnt = 1
    for i in btn:
        i.click()
        time.sleep(3)
        
        
        ## Checking if ith post is liked before or not
        cond = wait.until(EC.element_to_be_clickable((By.XPATH, "//span[contains(@class, 'xp7jhwk')]/div[contains(@role, 'button')]")))
        
        if(cond):    ## finding the like one
            # we are Useing JavaScript click because normal click doesn't work in this case
            driver.execute_script("arguments[0].click();", cond)
            time.sleep(2)
            
        else:
            previously_like = True
        
        
        ## exiting the post
        xpath = "//div[contains(@class, 'x160vmok x10l6tqk')]/div[@tabindex='0']"
        exit = driver.find_element(By.XPATH, xpath)
        exit.click()
        
        ## mainatining the number of post
        if cnt == no_of_post:
            break
        cnt += 1
        
    
    ## Closing profile
    close_profile(driver)
    
    return previously_like   
```

```python
## function to disliking the first no_of_post of a profile

def unlike(driver, profile, no_of_post):
    
    wait = WebDriverWait(driver, 10)       ## Explicit Wait
    
    previously_unlike = False
    
    ## opening profile
    search_open(driver, profile)
    
    ## Extracting The No_of_posts
    path = "//div[contains(@class,'x1lliihq x1n2onr6 xh8yej')]"
    btn = wait.until(EC.presence_of_all_elements_located((By.XPATH,path)))
    while len(btn) < no_of_post:
        driver.execute_script('window.scrollBy(0, 1000);')         ## Scrolling
        btn = wait.until(EC.presence_of_all_elements_located((By.XPATH,path)))
        
    print("number of posts extracted: ", len(btn))
    
    cnt = 1
    for i in btn:
        i.click()
        time.sleep(3)
        
        ## Checking if ith post is liked before or not
        cond = wait.until(EC.element_to_be_clickable((By.XPATH, "//span[contains(@class, 'xp7jhwk')]/div[contains(@role, 'button')]")))
        cond2 = wait.until(EC.element_to_be_clickable((By.XPATH, "//span[contains(@class, 'xp7jhwk')]/div[contains(@role, 'button')]/div//span//*")))
        
        ## finding the like one
        if(cond2.get_attribute("aria-label") == "Unlike"):
            # we are Useing JavaScript click because normal click doesn't work in this case
            driver.execute_script("arguments[0].click();", cond)
            time.sleep(2)
        else:
            previously_unlike = True
            
        ## exiting the post
        xpath = "//div[contains(@class, 'x160vmok x10l6tqk')]/div[@tabindex='0']"
        exit = driver.find_element(By.XPATH, xpath)
        exit.click()
        
        ## mainatining the number of post
        if no_of_post <= cnt:
            break
        cnt += 1
        
    ## coming back to home page
    close_profile(driver)
    
    return previously_unlike   
```

```python
## function to extract the username of no_of_followers folllowers of a profile

def extracting_followers_or_following(driver, profile, no_of_followers_following, followers_or_following):  
    
    time.sleep(3)
    
    fof = followers_or_following
    
    wait = WebDriverWait(driver, 20)       ## Explicit Wait
    
    # opening profile
    search_open(driver, profile)

    ## clicking on followers
    path = f'//li/a[contains(@href, "/{profile}/{fof}/")]'
    followers_following = wait.until(EC.element_to_be_clickable((By.XPATH, path)))
    followers_following.click()

    ## Extracting The No_of_followers_or_following
    path = '//div[@style= "height: auto; overflow: hidden auto;"]/div/div'
    btn = wait.until(EC.presence_of_all_elements_located((By.XPATH,path)))
    
    while len(btn) < no_of_followers_following:
        first_len_of_btn = len(btn)
        path = '//a[contains(@href, "/_foodiesm_/followers/")]/span[contains(@class, "_ac2a")]'
        scr1 = driver.find_element(By.XPATH, '//div[contains(@class, "_aano")]')                    ## followers window
        driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scr1)            ## Scrolling
        btn = wait.until(EC.presence_of_all_elements_located((By.XPATH,path)))
        
        time.sleep(5)

        driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scr1)            ## Scrolling
        btn = wait.until(EC.presence_of_all_elements_located((By.XPATH,path)))
        
        if len(btn) == first_len_of_btn:
            break

    print("number of followers extracted of", profile ,":", len(btn))
    

    path = '//a[contains(@class, "notranslate _a6hd")]//span'
    all_Username = wait.until(EC.presence_of_all_elements_located((By.XPATH, path)))
    
    ## making the list
    followers_followings = []
    cnt = 1
    for k in all_Username:
        followers_followings.append(k.text)
        if cnt == no_of_followers_followings:
            break
        cnt += 1
    
    print(followers_followings)
    
    ## exiting the followers page
    btn = driver.find_element(By.CLASS_NAME, "_abl-")
    btn.click()
    
    time.sleep(3)
    
    ## coming back to home page
    close_profile(driver)
        
    return followers_followings
```
